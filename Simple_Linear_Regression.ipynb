```python
import pandas as pd

dataset = pd.read_csv('/content/Salary Data.csv')
print(dataset.head ())

print(dataset.describe())

print(dataset.shape)
```

```python
import matplotlib.pyplot as plt
#plot the data

plt.scatter (dataset['YearsExperience'], dataset['Salary'])
plt.xlabel ('Years of Experience')
plt.ylabel ('Salary')
plt.title ('Salary vs Years of Experience')
plt.show()
```

```python
x= dataset.iloc[:, :-1]
y=dataset.iloc[:, -1]
```

```python
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_text, y_train, y_text = train_test_split(x, y, test_size=0.2, random_state=1)
print(x_train.shape)
print(x_text.shape)
print(y_train.shape)
print(y_text.shape)
```

```python
from sklearn.linear_model import LinearRegression

# create model

model = LinearRegression()

#train the model

model.fit(x_train, y_train)

y_pred = model.predict(x_text)
print(y_pred)
print(y_text)
```

```python
plt.scatter (x_train, y_train, color='red')
plt.plot(x_train, model.predict(x_train),color='blue')
plt.title('Salary vs experince (trianing set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter (x_text, y_text, color='red')
plt.plot(x_text, model.predict(x_text),color='blue')
plt.title('Salary vs experince (trianing set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print("mean Squared error: ",mean_squared_error(y_text,y_pred))
print("R2 Score: ",r2_score(y_text,y_pred))
```

```python
print('Slope: ',model.coef_)
print("Intercept: ", model.intercept_)
```

```python
years_of_experience =10
Salary = model.predict([[years_of_experience]])
print("Salary of someone with 10 years of experienc: ", Salary)
```
